```python
# Untuk membaca data
import pandas as pd
import re
```

### pre-processing kompas

```python
# Membaca JSON ke dalam DataFrame
df_kmps = pd.read_json('kompas_crawl_2024-03-01_to_2024-06-22.json')
df_kmps
```

```python
# Definisikan fungsi untuk membersihkan teks
def clean_text(text):
    # Definisikan pola RegEx untuk mencocokkan dan menghapus teks tidak penting
    pattern = r'\b(http\S+|baca juga|berita kompas com|kami langsung di|di ponselmu pilih|langsung di ponselmu|whatsapp|berita pilihan kami)\b.*?(\.|$)'
    
    # Hapus teks tidak penting menggunakan re.sub
    cleaned_text = re.sub(pattern, '', text, flags=re.IGNORECASE)
    
    # Hapus spasi berlebih yang mungkin ditinggalkan setelah penghapusan
    cleaned_text = ' '.join(cleaned_text.split())
    
    return cleaned_text

# Terapkan fungsi pembersihan ke kolom 'content'
df_kmps['content'] = df_kmps['content'].apply(clean_text)
```

### pre-processing Liputan6.com

```python
# Membaca JSON ke dalam DataFrame
df_lptn6 = pd.read_json('liputan6_crawl_2024-01-01_to_2024-06-23.json')
df_lptn6
```

```python
# Definisikan fungsi untuk membersihkan teks
def clean_text(text):
    # Definisikan pola RegEx untuk mencocokkan dan menghapus teks tidak penting
    pattern = r'\b(http\S+|baca juga|baca juga video|baca juga hasil)\b.*?(\.|$)'
    
    # Hapus teks tidak penting menggunakan re.sub
    cleaned_text = re.sub(pattern, '', text, flags=re.IGNORECASE)
    
    # Hapus spasi berlebih yang mungkin ditinggalkan setelah penghapusan
    cleaned_text = ' '.join(cleaned_text.split())
    
    return cleaned_text

# Terapkan fungsi pembersihan ke kolom 'content'
df_lptn6['content'] = df_lptn6['content'].apply(clean_text)
df_lptn6
```

```python
# Menggabungkan dua data frame secara vertikal
df = pd.concat([df_kmps, df_lptn6], ignore_index=True)
```

```python
# Simpan ke berkas JSON dengan orientasi 'records'
df.to_json('dt_berita.json', orient='records',indent=4)
```
